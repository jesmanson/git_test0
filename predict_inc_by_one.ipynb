```python
import numpy as np
import matplotlib.pyplot as plt

from keras.models import Sequential
from keras.layers import Dense, LSTM, BatchNormalization, Dropout, SimpleRNN
from keras.callbacks import EarlyStopping
```

```python
data1 = [i for i in range(100)]
data1 = np.array(data1)
data3 = [i for i in range(50,150)]
data3 = np.array(data3)
targets3 = [i for i in range(51,151)]
targets3 = np.array(targets3)
targets1 = [i for i in range(1,101)]
targets1 = np.array(targets1)
data2 = [i for i in range(200,300)]
data2 = np.array(data2)
targets2 = [i for i in range(201,301)]
targets2 = np.array(targets2)
```

```python
x_train = np.zeros((3,1,100)) #(nr of sequences, width, height)
x_train[0] = data1
x_train[1] = data2
x_train[2] = data3
y_train = np.zeros((3,1,100))
y_train[0]=targets1
y_train[1]=targets2
y_train[2]=targets3
```

```python
x_val = [i for i in range(100,200)]
x_val = np.array(x_val)
y_val = [i for i in range(101,201)]
y_val = np.array(y_val)
x_val = x_val.reshape((1,1,100))
y_val = y_val.reshape((1,1,100))
```

```python
model = Sequential()
model.add(SimpleRNN(100,input_shape=(1,100),return_sequences = True)) #return sequences => many->many nn
model.add(Dropout(0.2))
#model.add(BatchNormalization())
#model.add(LSTM(100, return_sequences=True))
#model.add(Dropout(0.2))
#model.add(LSTM(100,return_sequences=True))
#model.add(Dropout(0.2))
model.add(Dense(100))
model.compile(loss='mean_absolute_error',optimizer ='adagrad', metrics=['accuracy'])
```

```python
#estop = EarlyStopping(monitor='val_loss', patience=50)
model.fit(x_train,y_train,batch_size=1,verbose = 2, validation_data=(x_val,y_val),epochs=1000)#callbacks=[estop])
```

```python
x_test = [x for x in range(100,200)]
x_test = np.array(x_test)
x_test = x_test.reshape((1,1,100))
```
